```python
intents = {
    "greetings": {
        "patterns": ["hi", "hello", "hey", "hii"],
        "responses": ["Hey ", "Hello! ", "Hi there!"]
    },
    "price_query": {
        "patterns": ["price", "how much", "rate","cost"],
        "responses": ["Pls tell me the product name"]
    },
    "product_list": {
        "patterns": ["products", "list products", "available products"],
        "responses": ["Here's the list of available products:"]
    },
    "goodbye": {
        "patterns": ["bye", "see you", "take care"],
        "responses": ["Goodbye","See you later"]
    }
}

# Product prices
products = {
    "tomato": "₹40/kg",
    "onion": "₹30/kg",
    "potato": "₹20/kg",
    "carrot": "₹50/kg",
    "beetroot": "₹60/kg",
    "broccoli": "₹80/kg",
    "cabbage": "₹35/kg"
}
```

```python
import random
# Function to get intent
def get_intent(user_input):
    user_input = user_input.lower()
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return None

# Function to get response
def get_response(intent):
    if intent == "greetings":
        return random.choice(intents[intent]["responses"])
    else:
        return intents[intent]["responses"][0]

# Function to get product price
def get_product_price(product):
    return products.get(product.lower(), "Product not found!")

# Function to handle chatbot
def chatbot():
    while True:
        user_input = input("user: ")
        intent = get_intent(user_input)
        if intent == "price_query":
            print("Bot:", get_response(intent))
            user_input = input("user: ")
            price = get_product_price(user_input)
            if price != "Product not found!":
                print(f"Bot: {price}")
            else:
                print("Bot:", price)
        elif intent == "product_list":
            print("Bot:", get_response(intent))
            print("Bot: tomato, onion, potato, carrot, broccoli, cabbage")
        elif intent == "goodbye":
            print("Bot:", get_response(intent))
            break
        elif intent:
            print("Bot:", get_response(intent))
        else:
            print("Bot:", "Sorry, I didn’t get that!")

# Run chatbot
chatbot()
```
