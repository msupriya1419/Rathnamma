```python
def library_chatbot():
    print("Welcome to College Library Assistant Bot")

    # Sample library data
    books_db = {
        "data structures": 5,
        "operating systems": 6,
        "dbms": 3,
        "python programming": 2,
        "artificial intelligence": 0,
    }
    library_timings = "9 AM to 6 PM, Monday to Saturday"
    fine_per_day = 2

    while True:
        print("\nHow can I help you?")
        print("1.Issue a book 2.Return book & check fine 3.Library timings 4.Check availability")
        user_input = input().lower().strip()

        # if-else conditions
        if user_input == "exit":
            print("Library Assistant:Thank you Visit again!")
            break

        elif "1" in user_input or "issue" in user_input:
            book_name = input("Library Assistant: Which book do you want to issue?").lower().strip()

            # Check availability using if-else
            if book_name in books_db:
                if books_db[book_name] > 0:
                    books_db[book_name] -= 1
                    print(f"Library Assistant: '{book_name.title()}' issued successfully.")
                    print("Library Assistant: Return within 15 days to avoid fine.")
                else:
                    print(f"Library Assistant: Sorry,'{book_name.title()}'is currently out of stock.")
            else:
                print("Library Assistant: Sorry, we don't have that book")

        elif "2" in user_input or "return" in user_input or "fine" in user_input:
            # Exception handling
            try:
                days_late = int(input("Library Assistant: How many days late is the book? "))

            # Mathematical calculation
                if days_late > 0:
                    total_fine = days_late * fine_per_day
                    print(f"Library Assistant: Your fine is Rs. {total_fine}.")
                    print(f"Library Assistant: Calculation: {days_late} days * Rs.{fine_per_day} = Rs.{total_fine}")
                elif days_late ==0:
                    print("Library Assistant: No fine. Returned on time. Good job!")
                else:
                    print("Library Assistant: Days cannot be negative. Please enter valid input")

            except ValueError:
                print("Library Assistant: Invalid input! Please enter a number for days")

        elif "3" in user_input or "timing" in user_input:
            print(f"Library Assistant: Library timings are {library_timings}")

        elif "4" in user_input or "available" in user_input or "check" in user_input:
            book_name = input("Library Assistant: Which book availability to check? ").lower().strip()
            if book_name in books_db:
                print(f"Library Assistant: '{book_name.title()}' - {books_db[book_name]}copies available")
            else:
                print(f"Library Assistant: '{book_name.title()}' not found")

        else:
            print("Library Assistant: Sorry, I didn't understand. Please choose option 1-4 or type 'exit'.")

# Run the chatbot
library_chatbot()
```
